```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_hist(data, **kwargs):
	means = np.array([np.mean(x) for x in data])
	stds =  np.array([np.std(x) for x in data])
	m = np.min(means-1*stds)
	M = np.max(means+1*stds)

	return plt.hist(data, range=(m, M), **kwargs)
```

```python
trove_data = np.genfromtxt('with-trove.csv', delimiter=',') / 1e9
map_data = np.genfromtxt('with-map.csv', delimiter=',') / 1e9
```

```python
fig = plt.figure(figsize=(11,8))
ax = plt.subplot(1, 2, 1)
h = make_hist([map_data[100:,0], trove_data[100:,0]], label=('map', 'trove'))
plt.legend(loc=2)
plt.title('collect')
plt.xlabel('time [s]')
ax.xaxis.set_ticks(ax.get_xlim())

ax = plt.subplot(1, 2, 2)
h = make_hist([map_data[100:,1], trove_data[100:,1]], label=('map', 'trove'))
plt.legend()
plt.title('mediate')
ax.xaxis.set_ticks(ax.get_xlim())
plt.xlabel('time [s]')


plt.show()
```
